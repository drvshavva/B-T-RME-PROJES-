```python
num_images=1
num_channels=3
shapey=1536
shapex=2048
patchSize=128
nb_patch=int(shapex/patchSize)*int(shapey/patchSize); # total number of patches
print("patch sayısı:",nb_patch)
```

```python
def create_patches(img,k):
    a=np.zeros((num_images,num_channels,shapey,shapex),dtype=np.uint8)
    
    a[0,0,:,:] = img[:,:,0]#r kanalını a'ya atıyoruz
    a[0,1,:,:] = img[:,:,1]#g kanalı
    a[0,2,:,:] = img[:,:,2]# b kanalı
    set_x=a;
    
    patches_all=np.zeros(shape=(nb_patch,num_channels,patchSize,patchSize),dtype='uint8');
    
    first_lin_patch=0;
    first_col_patch=0; 
    for p in range(0,nb_patch): #for each patch        
        patch_prov=set_x[0,:,first_lin_patch : first_lin_patch + patchSize,first_col_patch : first_col_patch + patchSize]; # provisory patch (before possible rotation) 

    
        patches_all[p,:,:,:] = patch_prov
        first_col_patch += patchSize

    
        if first_col_patch==shapex: #run all collumns in a row, then pass to the next row
            first_col_patch=0
            first_lin_patch += patchSize

    set_x=patches_all; 
    for i in range(0,nb_patch):
        path=str(k)+'b_'+str(i)+'.tif'
        io.imsave('C:/Users/user/Desktop/128/Test/Normal/'+path,set_x[i])
    
```

```python
n=100
from skimage import io
import numpy as np
for i in range(1,n):
    path='b'+str(i)+'.tif'
    img=io.imread('D:/Bitirme/BACH ICIAR 2018/ICIAR2018_BACH_Challenge/Photos/Benign/'+path)
    create_patches(img,i)
```
